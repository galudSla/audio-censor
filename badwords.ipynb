```python
with open('/content/drive/MyDrive/en.txt', 'r') as en:
    bad_words = en.read().split()
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
bad_words_stemmed = [stemmer.stem(word) for word in bad_words]
```
